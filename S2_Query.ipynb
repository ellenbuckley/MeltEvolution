```python
import os,sys
import numpy as np
import tenacity
import logging
from sentinelsat import SentinelAPI, read_geojson, geojson_to_wkt
from datetime import date

logging.basicConfig(format='%(message)s', level='INFO')


username=''
passw=''
api = SentinelAPI(username, passw, 'https://scihub.copernicus.eu/dhus')

# download single scene by known product id
#api.download(<product_id>)
@tenacity.retry(stop=tenacity.stop_after_attempt(24), wait=
                (900))
def download_all(*args, **kwargs):
    return api.download_all(*args, **kwargs)

#polygon of region of interest
poly_path= '/Users/ebuckle2/Documents/Analysis/GIS/ROI2019.geojson'

# search by polygon, time, and SciHub query keywords
footprint = geojson_to_wkt(read_geojson(poly_path))
products = api.query(footprint,
                     date=('20190501', '20190801'),
                     platformname='Sentinel-2',
                     producttype='S2MSI1C',
                     cloudcoverpercentage=(0, 50))
#print(products)
products_df = api.to_dataframe(products)
downloaded, triggered, failed = download_all(products)
```
